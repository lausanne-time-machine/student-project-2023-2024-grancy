```python
job_count_dict_1923 = {}
for k in JOB_CAT_ABRV.keys():
    job_count_dict_1923[k] = 0

job_count_dict_1923['none'] = 0
```

```python
for index, row in df_1923.iterrows():
    job_in_row = row['cat_job']
    if type(job_in_row) is str:
        job_count_dict_1923[job_in_row] += 1
    else:
        job_count_dict_1923['none'] += 1
```

```python
job_count_dict_1951 = {}
for k in JOB_CAT_ABRV.keys():
    job_count_dict_1951[k] = 0

job_count_dict_1951['none'] = 0
```

```python
variations = {}

for k in job_count_dict_1951.keys():
    variations[k] = 0

variations['none'] = 0

for e in job_count_dict_1951.keys():
    variations[e] = job_count_dict_1951[e] - job_count_dict_1923[e]
```

```python
variations_per = {}

for k in job_count_dict_1951.keys():
    variations_per[k] = 0

variations_per['none'] = 0

for e in job_count_dict_1951.keys():
    if job_count_dict_1951[e] == 0:
        print(f"zero for {e}")
    else:
        variations_per[e] = (job_count_dict_1951[e] - job_count_dict_1923[e]) / job_count_dict_1951[e] * 100
```
